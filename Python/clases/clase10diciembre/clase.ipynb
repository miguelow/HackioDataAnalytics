```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None) 
```

```python
df = pd.read_csv("./movies_metadata.csv")
df.head()
```

```python
df.value_counts()
```

## 1 Producciones anuales

```python
print(df['release_date'].info())
print(df['release_date'].head())
```

```python
df['new_date'] = pd.to_datetime(df['release_date'], format='%Y-%m-%d', errors='coerce')
```

```python
print(df['new_date'].info())
print(df['new_date'].head())
```

```python
year_counts = df['new_date'].dt.year.value_counts().sort_index()

len(year_counts)
```

```python
sns.histplot(x=df['new_date'].dt.year, 
             data=df, 
             color="violet", 
             bins=len(year_counts))

plt.xlabel("Año")
plt.ylabel("Número de películas")
plt.title("Distribución de películas por año")
```

```python
plt.figure(figsize=(14, 8))
sns.lineplot(x=df['new_date'].dt.year, y=df['new_date'].dt.year, marker='o')
```
